```python
from scipy.optimize import differential_evolution
```

```python
PAINT_W = 7793
PAINT_H = 4409
RHO_S  = PAINT_W / PAINT_H
 
EPSILON   = 0.05
BIG_M     = 1e7 
```

```python
def dims(delta):
    w = 594 + 238 * delta   # δ=0 → 594 px (0°),  δ=1 → 832 px (90°)
    h = 832 - 238 * delta   # δ=0 → 832 px (0°),  δ=1 → 594 px (90°)
    return w, h

def objetivo(x):
    C     = int(round(x[0]))
    R     = int(round(x[1]))
    delta = int(round(x[2]))
 
    w, h = dims(delta)
    rho  = (C * w) / (R * h)
    err  = abs(rho - RHO_S) / RHO_S
 
    penal = BIG_M * max(0, err - EPSILON) ** 2
    return C * R + penal
```

```python
result = differential_evolution(
    objetivo,
    bounds      = [(1, 30), (1, 30), (0, 1)],  # C, R, δ
    integrality = [1, 1, 1],                   # all integers
    maxiter     = 5000,
    popsize     = 25,
    tol         = 1e-12,
    mutation    = (0.5, 1.5),
    recombination = 0.9,
)
```

```python
C     = int(round(result.x[0]))
R     = int(round(result.x[1]))
delta = int(round(result.x[2]))
w, h  = dims(delta)
 
N     = C * R
rho   = (C * w) / (R * h)
err   = abs(rho - RHO_S) / RHO_S
ori   = '90°' if delta else '0°'
 
print("───────────────────────────────────────────────────────────────")
print(f"  Painting:     {PAINT_W} × {PAINT_H} px  →  ρ* = {RHO_S:.6f}")
print()
print(f"  Rotation:     {ori}  (δ = {delta})")
print(f"  Grid:         C = {C} columns,  R = {R} rows")
print(f"  Best N:       {N} notebooks")
print(f"  Dimensions:   {C*w} × {R*h} px")
print(f"  ρ grid:       {rho:.6f}")
print(f"  Error ratio:  {err*100:.4f}%")
print()
print(f"  Converged:    {result.success}")
print(f"  Mensaje:      {result.message}")
print("───────────────────────────────────────────────────────────────")
 
```
